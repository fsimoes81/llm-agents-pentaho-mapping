```python
import os
from llama_index.core import SimpleDirectoryReader, VectorStoreIndex, SummaryIndex
from llama_index.core.node_parser import SentenceSplitter, SimpleNodeParser, TokenTextSplitter, TextSplitter
from llama_index.core.ingestion import IngestionPipeline
from llama_index.core.extractors import KeywordExtractor, TitleExtractor
from llama_index.core.schema import MetadataMode
from llama_index.llms.ollama import Ollama
import nest_asyncio
```

```python
pentaho_file_path = '../data/fact_sales.xml'
```

```python
documents = SimpleDirectoryReader(input_files=[pentaho_file_path]).load_data()
```

```python
parser = SimpleNodeParser.from_defaults()
nodes = parser.get_nodes_from_documents(documents)
```

```python
splitter = SentenceSplitter(chunk_size=1024, chunk_overlap=100)
pipeline = IngestionPipeline(transformations=[splitter])
nodes = pipeline.run(documents=documents)
```
